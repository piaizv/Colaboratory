```python
%%writefile porra/views.py

from django.http import HttpResponse

def index(request):
    return HttpResponse('Esta es la porra')

def detalle(request, iddetalle):
    return HttpResponse('detalle %s.' % iddetalle)
```

```python
%%writefile porra/urls.py

from django.urls import path

from . import views

urlpatterns = [
    path('', views.index, name='index'),
    path('porra/', views.index, name='porra'),
    path('detalle/<int:iddetalle>', views.detalle, name='detalle'),
]
```

```python
%%writefile djangoApp/urls.py

from django.contrib import admin
from django.urls import include, path

urlpatterns = [
    path('porra/', include('porra.urls')),
    #path('por', include('porra.urls')),
    path('admin/', admin.site.urls),
]
```
